```python
import pandas as pd
from itertools import combinations
```

```python
def jaccard_similarity(list1, list2):
    # Computa la similitud de Jaccard entre dos sets
    set1 = set(list1)
    set2 = set(list2)

    intersection = set1.intersection(set2)
    union = set1.union(set2)
    return len(intersection) / len(union)
```

```python
df = pd.read_csv('tweets_2022_abril_junio.csv') # cuidado al ejecutar
```

```python
df.drop(['id', 'created_at', 'retweet_count', 'favorite_count'], axis=1, inplace=True)
```

```python
df.shape
```

```python
tweets_originales = df.loc[df['text'].str[:4].str.contains('RT @',case=False) == False]

rt = df.loc[df['text'].str[:4].str.contains('RT @',case=False) == True]

retweets = pd.DataFrame()
split_text = rt['text'].str.split(' ', expand=True, n=2)
retweets['user'] = split_text[1].str.strip("@").str.strip(':')
retweets['tweet'] = split_text[2]

retweets.reset_index(drop=True, inplace=True)


tweets_originales.rename(columns={'screen_name': 'user', 'text': 'tweet'}, inplace=True)
tweets = pd.concat([tweets_originales, retweets]).reset_index(drop=True)


final_tweets = tweets.drop_duplicates(['user', 'tweet'], ignore_index=True)

final_tweets
```

```python
tweet_shingles = dict()
inverted_index = dict()
k=4

for id in range(final_tweets.shape[0]): #range(final_tweets.shape[0])  
    tweet_shingles[id] = set()
    
    for i in range(len(final_tweets['tweet'][id]) - k-1):        
        shingle = final_tweets['tweet'][id][i:i+k]
        tweet_shingles[id].add(shingle)

        if shingle not in inverted_index:
            inverted_index[shingle] = set()
        inverted_index[shingle].add(id)

    tweet_shingles[id] = list(tweet_shingles[id])

    if id%50000 == 0:
        print(f"{id} de 1450000")
```

```python
tweet_shingles
```

```python
len(inverted_index)
```

```python
filtered_inverted_index = set()
i = 0

for shingle, tweets in inverted_index.items():
    if len(tweets) <= 5:
        tweet_combinations = combinations(tweets, 2)
        for pair in tweet_combinations:
            filtered_inverted_index.add(tuple(sorted(pair)))

    i += 1

    if i%1000 == 0:
        print(f"{i} de 2000000")
```

```python
len(filtered_inverted_index)
```

```python
similar = []
s = 0.8

i = 0
for pair in list(filtered_inverted_index):
    tweet1 = tweet_shingles[pair[0]]
    tweet2 = tweet_shingles[pair[1]]

    sim = jaccard_similarity(tweet1, tweet2)
    if sim > s:
        similar.append((pair[0], pair[1]))

    
    if i%25000 == 0:
        print(f"{i} de 157000")

    i += 1
```

```python
len(similar)
```

```python
user_pairs = {}

i = 0

for pair in similar:
    if final_tweets["user"][pair[0]] != final_tweets["user"][pair[1]]:
        users = [final_tweets["user"][pair[0]],final_tweets["user"][pair[1]]]
        users = tuple(sorted(users))

        if users in user_pairs.keys():
            user_pairs[users] += 1
        else:
            user_pairs[users] = 1

    if i%500 == 0:
        print(f"{i} de 4600")

    i += 1
```

```python
user_pairs
```

```python
pares_finales = sorted(user_pairs.items(), key=lambda x: x[1], reverse=True)
```

```python
pares_finales
```

```python
for k,v in pares_finales:
    if v >= 3:
        print(f"Usuarios {k[0]} y {k[1]} tienen {v} tweets similares")
```
